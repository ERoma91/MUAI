## Ejercicio 5

En la clase MejorNumero, crea un nuevo método llamado iterar() que reciba una lista con 3 números romanos. A continuación, iterando sobre los elementos de la lista llamará a la función suma_romano(). Los posibles errores se tendrán que gestionar con excepciones (try...except...) para mostrar un mensaje y seguir ejecutando el siguiente número.

**NOTA:** el mensaje de error ha de seguir el siguiente formato --> print('Ha fallado el número', 5). 

```python
class MejorNumero(Numero):
    
    def __init__(self, normal):
        Numero.__init__(self, normal)
    

    def resta(self, romanoToRest):
        
        self.romanoToRest = romanoToRest
        self.normalToRest = self.convToNormal(romanoToRest)
        
        self.normal -= self.normalToRest  
        self.romano = self.convToRomano(self.normal)
        
    
    def multiplica(self, romanoToMul):
        
        self.romanoToMul = romanoToMul
        self.normalToMul = self.convToNormal(romanoToMul)
        
        self.normal *= self.normalToMul
        self.romano = self.convToRomano(self.normal)
    
    
    def iterar(self, argLista):
        
        self.argLista = argLista
        
        for i in range(len(argLista)):
            try:
                self.suma_romano(argLista[i])
            except:
                print("Ha fallado el número " + str(argLista[i]))
```
